```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import os, math, sys
import pandas as pd
from IPython.display import display
import sklearn.decomposition
```

```python
csv_directory = "/home/tushaar/Desktop/phishing_corpus/csv/"

files = ['dataset_HP', 'dataset_HS', 'dataset_HSP']
```

```python
'''
Read entire dataset ignoring class labels (40 dimensions)
'''

def get_data_no_class_labels(filename):
    filename = csv_directory + filename + '.csv'
    if os.path.exists(filename):
        df = pd.read_csv(filename)
        df = df.drop('label', 1)
    else:
        print "Run `feature_extraction.py` to get CSV"
    return df

filename = files[2]
no_class_label_df = get_data_no_class_labels(filename)
display(no_class_label_df.head())
```

```python
def PCA(no_class_label_df):
    df_np = df.as_matrix().astype(float)
    print "Dataframe without class labels: \n"
    print df_np
    print('==============================================\n\n')
    
    mean_vector = np.mean(df_np.T, axis = 1)
    print "Mean vector: \n"
    print mean_vector
    print('==============================================\n\n')
    
    center_cols = mean_vector - df_np
    print "Computing center columns: \n"
    print center_cols
    print('==============================================\n\n')
    
    cov_matrix = np.cov(center_cols.T)
    print "Covariance matrix: \n"
    print cov_matrix
    print('==============================================\n\n')
    
    eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
    print "Eigen values: \n"
    print eig_vals
    print('==============================================\n\n')
    print "Eigen vectors: \n"
    print eig_vecs
    
    projection = eig_vecs.T.dot(center_cols.T)
    return projection.T
    
projection = PCA(no_class_label_df)
```

```python
def sklearn_PCA(no_class_label_df):
    df_np = df.as_matrix().astype(float)
    print "Dataframe without class labels: \n"
    print df_np
    print('==============================================\n\n')
    
    pca = sklearn.decomposition.PCA()
    pca.fit(df_np)

    print "Eigen values: \n"
    print(pca.explained_variance_)
    print('==============================================\n\n')
    print "Eigen vectors: \n"
    print(pca.components_)

    sklearn_projection = pca.transform(df_np)
    return sklearn_projection
    
sklearn_projection = sklearn_PCA(no_class_label_df)
```

```python
def comparison(projection, sklearn_projection):
    print "projection: \n"
    print projection
    print('==============================================\n\n')
    
    print "sklearn.projection: \n"
    print sklearn_projection
    print('==============================================\n\n')
    
    difference = projection - sklearn_projection
    difference = np.around(difference)
    return difference

comparison(projection, sklearn_projection)
```
