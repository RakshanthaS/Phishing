```python
import numpy as np
import os
from IPython.display import display, Image
import subprocess
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
csv_directory = "/home/tushaar/Desktop/phishing_corpus/csv/"
feature_selection_csv_directory = "/home/tushaar/Desktop/phishing_corpus/feature_selection_csv/"
decision_trees_directory = "/home/tushaar/Desktop/phishing_corpus/decision_trees/"
```

```python
def get_data(filename):
    filename = feature_selection_csv_directory + filename + '.csv'
    if os.path.exists(filename):
        df = pd.read_csv(filename)
    else:
        print "Run `feature_extraction.py` to get CSV"
    return df

filename = 'dataset_HP'
df = get_data(filename)
display(df.head())
display(df.tail())
```

```python
def attribute_types(df, attribute):
    return df[attribute].unique()

def preprocessing(df, label_column):
    transformed_df = df.copy()
    labels = attribute_types(df, label_column)
    mapping = {name : number for number, name in enumerate(labels)}
    transformed_df[label_column] = transformed_df[label_column].replace(mapping)
    return mapping, transformed_df

mapping, transformed_df = preprocessing(df, 'label')
display(transformed_df.head())
display(mapping)
```

```python
def get_attributes(df):
    attributes = list(df.columns[:-1])
    return attributes

attributes = get_attributes(transformed_df)
display(attributes)
print len(attributes)
```

```python
def fitting_decision_tree(df, attributes, target_column):
    y = df[target_column]
    X = df[attributes]
    decision_tree = DecisionTreeClassifier(criterion = "entropy", random_state = 99)
    decision_tree.fit(X, y)
    return decision_tree

decision_tree = fitting_decision_tree(transformed_df, attributes, 'label')
print decision_tree
```

```python
def view_decision_tree(filename, decision_tree, attributes):
    dot_file = decision_trees_directory + filename + '.dot'
    png_file = decision_trees_directory + filename + '.png'
    with open(dot_file, 'w') as output:
        export_graphviz(decision_tree, out_file = output, feature_names = attributes)
    
    convert_to_png = ["dot", "-Tpng", dot_file, "-o", png_file]
    try:
        subprocess.check_call(convert_to_png)
        os.remove(dot_file)
        print "Decision Tree creation: Success"
    except:
        print "Could not generate `png` image"

view_decision_tree(filename, decision_tree, attributes)
```

```python
def external_testing(df, label):
    print "Extracting features"
    features = get_attributes(df)
    y = df[label].as_matrix().astype(float)
    X = df[features].as_matrix().astype(float)
    
    print "Splitting into training and testing sets"
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)
    
    print "Classify using CART"
    estimator = DecisionTreeClassifier(max_leaf_nodes=3, random_state=0)
    estimator.fit(X_train, y_train)
    y_score = estimator.predict(X_test)
    
    mis_classifications = 0
    for index in range(len(y_score)):
        # print "Predicted, Actual = (%s, %s)" %(y_score[index], y_test[index])
        if y_score[index] != y_test[index]:
            mis_classifications = mis_classifications + 1
            print "Input: %s" %X_test[index]
            print "Predicted, Actual = (%s, %s)\n" %(y_score[index], y_test[index])
    print "Number of mis-classifications = %s (out of %s)" %(mis_classifications, len(y_test))

external_testing(transformed_df, 'label')
```
